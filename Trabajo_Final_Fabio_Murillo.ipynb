```python
df.count()
```

```python
df["Fecha"] = pd.to_datetime(df["Fecha"])
```

```python
df.head()
```

```python
locator = Nominatim(user_agent='FabiosGeocoder')
```

```python
location = locator.geocode('San Jose, Costa Rica')
```

```python
# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=0.5)
```

```python
n = 4000

list_df = np.array_split(df, n)


for i in range(0, n):
    print (i)
    start = time.time()
    # 2- - create location column
    list_df[i]['localizacion'] = list_df[i]['Canton'].apply(geocode)
    # 3 - create longitude, latitude and altitude from location column (returns tuple)
    list_df[i]['punto'] = list_df[i]['localizacion'].apply(lambda loc: tuple(loc.point) if loc else None)
    # 4 - split point column into latitude, longitude and altitude columns and remove empty points
    list_df[i] = list_df[i][list_df[i]['punto'].notnull()]
    list_df[i][['latitud', 'longitud', 'altitud']] = pd.DataFrame(list_df[i]['punto'].tolist(), index=list_df[i].index)

    
    end = time.time()
    print((end - start)/60)

```

```python

dfconsol = pd.DataFrame()

for x in list_df:
    frames = [dfconsol, x]
    dfconsol = pd.concat(frames)
    
   
```
